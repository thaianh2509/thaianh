```python
consumer = df.loc[df['Segment']=="Consumer"]
consumer.head()
```

```python
consumer_profit = pd.pivot_table(consumer, values='Profit',index='State',
                    columns=['year'], aggfunc=np.sum, fill_value=0).reset_index()
consumer_profit['total'] = consumer_profit[2014]+consumer_profit[2015]+consumer_profit[2016]+consumer_profit[2017]

consumer_profit.sort_values(by= 'total',ascending=True, inplace=True)
consumer_profit.head(10)
```

```python
fig, (ax1) = plt.subplots(nrows=1, ncols=1,figsize=(30,15)) 


xs = consumer_profit['State']
ys = consumer_profit['total']


ax1.bar(xs,ys)
for x,y in zip(xs,ys):

        label = "{:.2f}".format(y)

        plt.annotate(label, 
                    (x,y), 
                    textcoords="offset points", 
                    fontsize= 10,
                    xytext=(0,10),
                    ha='center') 

        
                    
plt.xlabel("State")
plt.ylabel("profit")
plt.xticks(consumer_profit['State'],rotation='vertical',fontsize= 20 )
ax1.title.set_text('Li nhun Consumer theo cc bang')
plt.grid()
plt.show()
```

```python
#Cc sn phm h mua l g? S lng bao nhiu?
```

```python
consumer_subCategory = pd.pivot_table(consumer, values='Quantity',index='Sub-Category',
                    columns=['year'], aggfunc=np.sum, fill_value=0).reset_index()
consumer_subCategory['total'] = consumer_subCategory[2014]+consumer_subCategory[2015]+consumer_subCategory[2016]+consumer_subCategory[2017]

consumer_subCategory.sort_values(by= 'total',ascending=False, inplace=True)
consumer_subCategory.head(10)
```

```python
fig, (ax1) = plt.subplots(nrows=1, ncols=1,figsize=(30,15)) 


xs = consumer_subCategory['Sub-Category']
ys = consumer_subCategory['total']


ax1.bar(xs,ys)
for x,y in zip(xs,ys):

        label = "{:.2f}".format(y)

        plt.annotate(label, 
                    (x,y), 
                    textcoords="offset points", 
                    fontsize= 12,
                    xytext=(0,10), 
                    ha='center') 
                    
plt.xlabel("Category")
plt.ylabel("S lng")
plt.xticks(consumer_subCategory['Sub-Category'],rotation=40,fontsize= 20 )
ax1.title.set_text('S lng theo Category')
plt.grid()
plt.show()
```
